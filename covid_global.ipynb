Cek Data dari perbedadaan banyak data antara dfc dan dfr

index ke berapa ada data yang tidak sama di dfc dan dfr?

```
list_beda=[] #menyimpan index yang datanya tidak sama antara dfc dan dfr
def cekdata(a,b):
  for i in range(len(b)):
    if (a.iloc[i]["Country/Region"]) != (b.iloc[i]["Country/Region"]):
         list_beda.append(i)
  return list_beda
cek = cekdata(dfc,dfr)
cek
```

```
list_cek=[]
all_cek=[]
for c in range(len(cek)):
  list_cek=dfc.iloc[[cek[c]]]["Country/Region"].values.tolist(),dfr.iloc[[cek[c]]]["Country/Region"].values.tolist()
  all_cek.append(list_cek)
all_cek
```
